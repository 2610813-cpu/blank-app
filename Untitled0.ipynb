```python
import requests

# 1. 한반도 상공의 위도/경도 범위 상자(Bounding Box) 설정
lamin = 33.0  # 최소 위도 (제주도 남쪽)
lamax = 39.0  # 최대 위도 (휴전선 부근)
lomin = 124.0 # 최소 경도 (서해)
lomax = 132.0 # 최대 경도 (동해)

# 2. OpenSky API 주소 (우리가 설정한 위경도 변수를 넣어줍니다)
url = f"https://opensky-network.org/api/states/all?lamin={lamin}&lomin={lomin}&lamax={lamax}&lomax={lomax}"

print("한반도 상공의 비행기 데이터를 요청합니다...")

# 3. 인터넷을 통해 서버에 데이터 요청하기
response = requests.get(url)

# 4. 응답이 성공적(200)인지 확인하고 데이터 열어보기
if response.status_code == 200:
    data = response.json()       # 받아온 데이터를 파이썬 딕셔너리로 변환
    states = data.get('states')  # 'states'라는 이름표가 붙은 비행기 목록 꺼내기

    if states:
        print(f"현재 한반도 상공에서 {len(states)}대의 비행기를 찾았습니다!\n")

        # 데이터가 너무 많으니 처음 5대만 예시로 출력해볼게요
        for state in states[:5]:
            # OpenSky 데이터는 리스트 형태라 숫자로 순서를 찾아야 합니다.
            callsign = state[1].strip() if state[1] else "알 수 없음" # 항공편명
            country = state[2]  # 국적
            lon = state[5]      # 경도
            lat = state[6]      # 위도
            alt = state[7]      # 고도 (미터)

            print(f"✈️ 항공편: {callsign} | 국적: {country} | 위치: ({lat}, {lon}) | 고도: {alt}m")
    else:
        print("현재 조건에 맞는 비행기가 없거나 일시적인 오류입니다.")
else:
    print(f"데이터를 가져오는 데 실패했습니다. 에러 코드: {response.status_code}")
```

```python
import requests # 웹 서버(API)와 인터넷 통신을 주고받기 위해 requests 라이브러리를 불러옵니다.
import pandas as pd # 받아온 복잡한 리스트 데이터를 다루기 쉬운 표(데이터프레임)로 바꾸기 위해 pandas를 불러옵니다.

def get_korea_flights_oauth2(): # 전체 과정을 하나의 함수로 묶어 나중에 앱에서 쉽게 호출할 수 있게 합니다.

    # [1단계: OAuth2 인증 토큰 받기]
    token_url = "https://opensky-network.org/oauth/token" # 토큰을 발급해 주는 서버의 주소를 지정합니다. (OAuth2 가정)
    client_id = "YOUR_CLIENT_ID" # API 제공자로부터 발급받은 내 앱의 고유 ID를 설정합니다.
    client_secret = "YOUR_CLIENT_SECRET" # 내 앱의 비밀번호 역할을 하는 시크릿 키를 설정합니다.

    try: # 토큰 발급 중 인터넷이 끊기거나 서버가 죽는 에러에 대비해 try 블록으로 감쌉니다.
        # OAuth2의 Client Credentials 방식으로 토큰을 요청하기 위해 필요한 정보들을 딕셔너리로 만듭니다.
        token_data = {'grant_type': 'client_credentials', 'client_id': client_id, 'client_secret': client_secret}

        # 토큰 발급 서버로 데이터를 보내며(POST), 서버가 느릴 수 있으니 30초까지만 기다리도록(timeout) 설정합니다.
        token_response = requests.post(token_url, data=token_data, timeout=30)

        # 응답 상태가 200번대(성공)가 아니면 코드를 멈추고 예외를 발생시키도록 강제합니다.
        token_response.raise_for_status()

        # 서버가 보내준 JSON 응답에서 접근 권한을 증명하는 'access_token' 글자만 뽑아내어 저장합니다.
        access_token = token_response.json().get('access_token')

    except requests.exceptions.Timeout: # 30초 동안 서버가 응답하지 않을 때 이 블록이 실행됩니다.
        print("에러: 토큰 발급 요청 시간이 초과되었습니다 (30초).") # 사용자에게 타임아웃 발생을 알립니다.
        return None # 토큰이 없으면 다음 단계를 진행할 수 없으니 함수를 종료합니다.
    except requests.exceptions.RequestException as e: # 비밀번호가 틀렸거나 인터넷이 끊기는 등 기타 모든 통신 에러를 잡습니다.
        print(f"에러: 토큰 발급에 실패했습니다 - {e}") # 정확히 어떤 에러가 났는지 원인을 출력합니다.
        return None # 마찬가지로 에러가 났으니 함수를 즉시 종료합니다.

    # [2단계: 발급받은 토큰으로 한반도 상공 비행기 데이터 조회하기]
    # 한반도 영역(위도 33~39, 경도 124~132)의 비행기만 걸러서 달라고 API 주소에 변수(쿼리 파라미터)를 포함시킵니다.
    api_url = "https://opensky-network.org/api/states/all?lamin=33.0&lomin=124.0&lamax=39.0&lomax=132.0"

    # 서버 문지기에게 나를 증명하기 위해, 아까 받은 토큰을 'Bearer'라는 표준 약속된 형태로 헤더(봉투)에 담습니다.
    headers = {'Authorization': f'Bearer {access_token}'}

    try: # 데이터를 받아오는 과정에서도 서버 지연이나 통신 에러가 생길 수 있으니 다시 try 블록을 씁니다.
        # 만들어둔 주소로 정보 요청(GET)을 보내고, 인증 봉투(headers)를 건네주며, 역시 30초만 기다립니다.
        response = requests.get(api_url, headers=headers, timeout=30)

        # API 권한이 없거나(401) 잘못된 요청(400)일 경우 코드를 멈추고 에러를 발생시킵니다.
        response.raise_for_status()

        # 성공적으로 받은 비행기 데이터 텍스트(JSON)를 파이썬이 다루기 쉬운 딕셔너리로 변환해 저장합니다.
        data = response.json()

    except requests.exceptions.Timeout: # 데이터 요청 시 30초가 넘어갈 경우를 대비한 예외 처리입니다.
        print("에러: 비행기 데이터 요청 시간이 초과되었습니다 (30초).") # 화면에 너무 오래 걸렸다고 알려줍니다.
        return None # 데이터를 못 받았으니 함수를 종료합니다.
    except requests.exceptions.RequestException as e: # 인증 실패 등 기타 이유로 데이터를 못 받아왔을 때의 처리입니다.
        print(f"에러: 비행기 데이터를 가져오는 데 실패했습니다 - {e}") # 에러 메시지를 출력합니다.
        return None # 함수를 종료합니다.

    # [3단계: 결과를 pandas 데이터프레임으로 변환]
    # OpenSky API가 주는 데이터는 이름표 없이 값만 덩그러니 있어서, 순서대로 무슨 데이터인지 이름표(컬럼명)를 정의합니다.
    columns = ['icao24', 'callsign', 'origin_country', 'time_position', 'last_contact', 'longitude', 'latitude', 'baro_altitude', 'on_ground', 'velocity', 'true_track', 'vertical_rate', 'sensors', 'geo_altitude', 'squawk', 'spi', 'position_source']

    # 받아온 데이터(data)가 비어있지 않고, 그 안에 'states'(실제 비행기 리스트)라는 값이 존재하는지 안전하게 확인합니다.
    if data and data.get('states'):
        # 이름표 없는 비행기 리스트(data['states'])에 우리가 만든 이름표(columns)를 붙여서 엑셀 같은 표(데이터프레임)로 만듭니다.
        df = pd.DataFrame(data['states'], columns=columns)
        return df # 표로 예쁘게 정리된 데이터프레임을 함수의 결과값으로 반환합니다.
    else: # 새벽 시간 등 조건에 맞는 비행기가 한 대도 없을 수 있기 때문에 예외 상황을 처리합니다.
        print("현재 한반도 상공에 조회되는 비행기가 없습니다.") # 상황을 사용자에게 알립니다.
        # 웹앱(Streamlit)에서 에러가 나지 않도록, 내용물은 없고 컬럼(이름표)만 있는 빈 데이터프레임을 반환합니다.
        return pd.DataFrame(columns=columns)

# 함수 실행 테스트용 코드입니다.
# df = get_korea_flights_oauth2()
# if df is not None:
#     print(df.head()) # 데이터가 잘 표로 만들어졌는지 위에서부터 5줄만 출력해봅니다.
```

```python
import pandas as pd
import numpy as np # 결측치(NaN)를 만들기 위해 넘파이 라이브러리 사용

# 1. 테스트용 가상 데이터 만들기 (API에서 받아왔다고 상상해 보세요!)
mock_data = {
    'icao24': ['111111', '222222', '333333', '444444', '555555'],
    'callsign': ['KOR123', 'ASI456', 'JEJ789', 'JIN012', 'AIR345'],
    'origin_country': ['South Korea'] * 5,
    'latitude': [37.5, 33.5, np.nan, 35.1, 36.5],    # 3번째 비행기는 위도 누락 (결측치)
    'longitude': [126.9, 126.5, 128.0, 129.0, np.nan], # 5번째 비행기는 경도 누락 (결측치)
    'on_ground': [False, True, False, False, False],   # 2번째 비행기는 지상(공항)에 있음
    'vertical_rate': [0.0, 0.0, 2.5, np.nan, -1.5]     # 4번째 비행기는 수직속도 누락 (결측치)
}

# 딕셔너리를 판다스 데이터프레임 'df'로 변환
df = pd.DataFrame(mock_data)

print("=== 0. 우리가 만든 원본 데이터 ===")
print(df)
print("\n" + "="*40 + "\n")


# ---------------------------------------------------------
# 여기서부터는 아까 작성했던 전처리 코드입니다. 이제 정상 작동할 거예요!
# ---------------------------------------------------------

print("=== 1. 전처리 전 데이터 상태 확인 ===")
initial_count = len(df)
print(f"총 비행기 대수: {initial_count}대")
print("\n[각 컬럼별 결측치 개수]")
print(df.isnull().sum())


print("\n=== 2. 지상에 있는 비행기(on_ground=True) 제외 ===")
df_flying = df[df['on_ground'] == False]
print(f"비행 중인 비행기 대수: {len(df_flying)}대")


print("\n=== 3. 필수 데이터(위도, 경도, 수직속도) 결측치 행 제거 ===")
df_clean = df_flying.dropna(subset=['latitude', 'longitude', 'vertical_rate'])


print("\n=== 4. 전처리 전후 데이터 개수 비교 ===")
final_count = len(df_clean)
print(f"전처리 전 비행기: {initial_count}대")
print(f"전처리 후 비행기: {final_count}대")
print(f"총 {initial_count - final_count}대의 데이터가 제거되었습니다.")

print("\n=== 5. 최종 완성된 깨끗한 데이터 ===")
print(df_clean)
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import requests
import pydeck as pdk

# 1. 웹앱 기본 설정 (가로로 넓게 쓰기)
st.set_page_config(page_title="실시간 항공 레이더", layout="wide")
st.title("✈️ 실시간 한반도 비행기 레이더")

# 2. 데이터 가져오기 함수 (캐싱 적용: 너무 자주 새로고침되어 차단당하는 것 방지)
@st.cache_data(ttl=60) # 60초 동안은 기존 데이터를 기억해둡니다.
def get_flight_data():
    url = "https://opensky-network.org/api/states/all?lamin=33.0&lomin=124.0&lamax=39.0&lomax=132.0"
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        data = response.json()

        if data and data.get('states'):
            columns = ['icao24', 'callsign', 'origin_country', 'time_position', 'last_contact',
                       'longitude', 'latitude', 'baro_altitude', 'on_ground', 'velocity',
                       'true_track', 'vertical_rate', 'sensors', 'geo_altitude', 'squawk', 'spi', 'position_source']
            df = pd.DataFrame(data['states'], columns=columns)

            # [전처리] 지상 비행기 제외 및 결측치 제거
            df = df[df['on_ground'] == False]
            df = df.dropna(subset=['latitude', 'longitude', 'baro_altitude'])

            # callsign(항공편명)이 비어있는 경우 'Unknown'으로 처리
            df['callsign'] = df['callsign'].apply(lambda x: x.strip() if isinstance(x, str) and x.strip() else "Unknown")
            return df
    except Exception as e:
        st.error(f"데이터를 가져오는 중 오류가 발생했습니다: {e}")
        return None
    return pd.DataFrame() # 데이터가 없을 경우 빈 데이터프레임 반환

# 3. 데이터 로드 및 화면 표시
with st.spinner("하늘에 있는 비행기들을 찾고 있습니다..."):
    df = get_flight_data()

if df is not None and not df.empty:
    st.success(f"현재 한반도 상공에 **{len(df)}대**의 비행기가 날고 있습니다!")

    # 4. Pydeck 3D 지도 그리기
    # 비행기를 점(Scatterplot)으로 표시
    layer = pdk.Layer(
        "ScatterplotLayer",
        data=df,
        get_position="[longitude, latitude]", # 위치 (경도, 위도 순서주의!)
        get_radius=3000,                      # 점의 크기 (반경 3km)
        get_fill_color="[255, 75, 75, 200]",  # 점의 색상 (빨간색, 투명도)
        pickable=True,                        # 마우스를 올렸을 때 반응하도록 설정
    )

    # 지도의 초기 중심점과 시야각 설정 (한반도 중심)
    view_state = pdk.ViewState(
        latitude=36.0,
        longitude=127.5,
        zoom=5.5,
        pitch=45, # 45도 기울여서 3D 느낌 내기
    )

    # 지도 출력
    st.pydeck_chart(pdk.Deck(
        layers=[layer],
        initial_view_state=view_state,
        map_style="mapbox://styles/mapbox/dark-v10", # 멋진 다크 모드 지도
        tooltip={"text": "항공편: {callsign}\n고도: {baro_altitude}m\n국적: {origin_country}"} # 마우스 오버 시 정보
    ))

    # 새로고침 버튼
    if st.button("🔄 데이터 새로고침"):
        st.cache_data.clear()
        st.rerun()
else:
    st.warning("현재 표시할 비행기 데이터가 없습니다. 잠시 후 다시 시도해주세요.")
```
